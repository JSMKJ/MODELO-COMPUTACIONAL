### Ejercicio 1 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} k
$$

#### Solución matemática

Se utiliza la fórmula de la suma de los primeros \( n \) números naturales:

$$
\sum_{k=1}^{n} k = \frac{n(n+1)}{2}
$$

#### Solución en Python


```python
def suma_naturales(n):
    return n * (n + 1) // 2

print(suma_naturales(10))
```

---

### Ejercicio 2 (resuelto)

Calcule la siguiente sumatoria:

$$
\sum_{k=1}^{n} (2k - 1)
$$

#### Solución matemática

Esta expresión corresponde a la suma de los primeros \( n \) números impares:

$$
\sum_{k=1}^{n} (2k - 1) = n^2
$$

#### Solución en Python


```python
def suma_impares(n):
    return sum(2*k - 1 for k in range(1, n + 1))

print(suma_impares(10))
```

```python
# Ejercicio 3: suma de cuadrados
def suma_cuadrados(n):
    return sum(k**2 for k in range(1, n + 1))

n = 5
print(f"Ejercicio 3: Suma de cuadrados (n={n})")
print(f"Resultado: {suma_cuadrados(n)}")
```

## II. Ejercicios de Diccionarios en Python

```python
# Diccionario 1: Estudiantes y Calificaciones
estudiantes = {
    "Ana": 95,
    "Bruno": 87,
    "Carlos": 92,
    "Diana": 88,
    "Enrique": 90
}
print("Diccionario 1: Estudiantes y Calificaciones")
print(f"Estudiantes: {estudiantes}")
promedio = sum(estudiantes.values()) / len(estudiantes)
print(f"Promedio: {promedio:.2f}")
```

```python
# Diccionario 2: Incrementar precios en 10%
productos = {
    "manzana": 1.50,
    "naranja": 2.00,
    "plátano": 0.80,
    "uva": 3.50
}
print("Diccionario 2: Incrementar precios en 10%")
print(f"Precios originales: {productos}")
productos_aumentados = {k: v * 1.10 for k, v in productos.items()}
print(f"Precios con aumento: {productos_aumentados}")
```

```python
# Diccionario 3: Encontrar la clave del valor máximo
numeros = {"a": 45, "b": 23, "c": 89, "d": 12, "e": 78}
print("Diccionario 3: Valor máximo")
print(f"Diccionario: {numeros}")
clave_max = max(numeros, key=numeros.get)
print(f"Clave con valor máximo: {clave_max} ({numeros[clave_max]})")
```

```python
# Diccionario 4: Invertir claves y valores
original = {"x": 10, "y": 20, "z": 30}
print("Diccionario 4: Invertir claves y valores")
print(f"Original: {original}")
invertido = {v: k for k, v in original.items()}
print(f"Invertido: {invertido}")
```

```python
# Diccionario 5: Eliminar palabras con frecuencia < 3
palabras_freq = {
    "python": 5,
    "hola": 2,
    "código": 4,
    "test": 1,
    "función": 3,
    "bug": 2
}
print("Diccionario 5: Filtrar por frecuencia")
print(f"Original: {palabras_freq}")
palabras_filtradas = {k: v for k, v in palabras_freq.items() if v >= 3}
print(f"Filtrado (freq >= 3): {palabras_filtradas}")
```

```python
# Diccionario 6: Combinar dos diccionarios
dict1 = {"a": 10, "b": 20, "c": 30}
dict2 = {"a": 5, "b": 15, "d": 25}
print("Diccionario 6: Combinar diccionarios")
print(f"Diccionario 1: {dict1}")
print(f"Diccionario 2: {dict2}")
combinado = dict1.copy()
for k, v in dict2.items():
    combinado[k] = combinado.get(k, 0) + v
print(f"Combinado: {combinado}")
```

```python
# Diccionario 7: Ordenar diccionario por valores (ascendente)
calificaciones = {"Carlos": 85, "Ana": 92, "Bruno": 78, "Diana": 88}
print("Diccionario 7: Ordenar por valores")
print(f"Original: {calificaciones}")
ordenado = dict(sorted(calificaciones.items(), key=lambda x: x[1]))
print(f"Ordenado: {ordenado}")
```

```python
# Diccionario 8: Verificar si una clave existe
inventario = {"laptop": 5, "mouse": 20, "teclado": 15}
print("Diccionario 8: Verificar existencia de clave")
print(f"Inventario: {inventario}")
clave_buscar = "mouse"
if clave_buscar in inventario:
    print(f"'{clave_buscar}' SÍ existe en el diccionario")
else:
    print(f"'{clave_buscar}' NO existe en el diccionario")
```

## III. Ejercicios de Multiplicación de Diccionarios en Python

(Se entiende por multiplicación la operación valor a valor entre diccionarios con claves comunes).

```python
# Multiplicación 1: Producto de valores correspondientes
dict_a = {"x": 2, "y": 3, "z": 4}
dict_b = {"x": 5, "y": 6, "z": 7}
print("Multiplicación 1: Producto de valores correspondientes")
print(f"Diccionario A: {dict_a}")
print(f"Diccionario B: {dict_b}")
producto = {k: dict_a[k] * dict_b[k] for k in dict_a}

print(f"Producto: {producto}")
```

```python
# Multiplicación 2: Multiplicar diccionario por un escalar
datos = {"a": 10, "b": 20, "c": 30}
escalar = 2.5
print("Multiplicación 2: Multiplicar por un escalar")
print(f"Diccionario: {datos}")
print(f"Escalar: {escalar}")
multiplicado = {k: v * escalar for k, v in datos.items()}

print(f"Resultado: {multiplicado}")
```

```python
# Multiplicación 3: Producto punto de dos vectores
vector_v = {"i": 1, "j": 2, "k": 3}
vector_w = {"i": 4, "j": 5, "k": 6}
print("Multiplicación 3: Producto punto de dos vectores")
print(f"Vector V: {vector_v}")
print(f"Vector W: {vector_w}")
producto_punto = sum(vector_v[k] * vector_w[k] for k in vector_v)
print(f"Producto punto (V · W): {producto_punto}")
```

```python
# Multiplicación 4: Costo total por producto y general
precios = {
    "laptop": 800,
    "mouse": 25,
    "teclado": 75,
    "monitor": 300
}
cantidades = {
    "laptop": 2,
    "mouse": 5,
    "teclado": 3,
    "monitor": 1
}
print("Multiplicación 4: Costo total por producto y general")
print(f"Precios: {precios}")
print(f"Cantidades: {cantidades}")
costo_por_producto = {k: precios[k] * cantidades[k] for k in precios}
costo_total = sum(costo_por_producto.values())
print(f"Costo por producto: {costo_por_producto}")
print(f"Costo total general: ${costo_total}")
```
